```
install.packages("SportsAnalytics") # NBA1516各球員的數據
```

```
library(SportsAnalytics)
NBA1516<-fetch_NBAPlayerStatistics("15-16")
```

```
NBA1516OrderM<-NBA1516[order(NBA1516$TotalMinutesPlayed,decreasing=T),]
NBA1516OrderM[1:20,"Name"]          #出場分鐘數最高的前20名選手的名字,由大至小
```

```
subset(NBA1516,Team=="GSW")  #找出勇士球員
```

```
NBA1516[grepl("Curry",NBA1516$Name),]   #找出Curry的球員
```

```
library(ggplot2)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516) #製作散佈圖 qplot(x軸名稱,y軸名稱,data=使用資料)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,geom=c("point","smooth"))    #在圖上加漸進線
```

```
qplot(TotalPoints,data=NBA1516,fill=Position)    #如果輸入的變量並非雙變量,而是單變量時,預設圖形會從 散佈圖 → 直方圖
```
