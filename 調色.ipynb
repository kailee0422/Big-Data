```
install.packages("SportsAnalytics") # NBA1516各球員的數據
```

```
library(SportsAnalytics)
NBA1516<-fetch_NBAPlayerStatistics("15-16")
```

```
NBA1516OrderM<-NBA1516[order(NBA1516$TotalMinutesPlayed,decreasing=T),]
NBA1516OrderM[1:20,"Name"]          #出場分鐘數最高的前20名選手的名字,由大至小
```

```

install.packages("ggplot2")
```

```
library(ggplot2)
df <-data.frame(gp=factor(rep(letters[1:3],each=10)),y=rnorm(30))
ggplot(df,aes(x = gp,y=y))+geom_point(color="#ffa600")
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516) +geom_point(color="#003f5c")#製作散佈圖 qplot(x軸名稱,y軸名稱,data=使用資料)
```

```
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,geom=c("point","smooth"))+geom_line(color="#444e86")
```
